```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h",
                                                                    "price_change_percentage_7d",
                                                                    "price_change_percentage_14d",
                                                                    "price_change_percentage_30d",
                                                                    "price_change_percentage_60d",
                                                                    "price_change_percentage_200d",
                                                                    "price_change_percentage_1y"]])
```

```python
# Create a DataFrame with the scaled data
df_market_data_transformed = pd.DataFrame(market_data_scaled, columns = ["price_change_percentage_24h",
                                                                        "price_change_percentage_7d",
                                                                        "price_change_percentage_14d",
                                                                        "price_change_percentage_30d",
                                                                        "price_change_percentage_60d",
                                                                        "price_change_percentage_200d",
                                                                        "price_change_percentage_1y"])

# Copy the crypto names from the original DataFrame
df_market_data_transformed["coin_id"] = df_market_data.index

# Set the coin_id column as index
df_market_data_transformed = df_market_data_transformed.set_index('coin_id')

# Display the scaled DataFrame
df_market_data_transformed.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0, n_init='auto')
    k_model.fit(df_market_data_transformed)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_curve_original = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
elbow_curve_original
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Based on the elbow method, the optimal number of clusters (k) is 4.

I plotted the inertia (within-cluster sum of squares) against various values of k and observed that the curve shows a clear "elbow" at k=4. This indicates a point of diminishing returns, where increasing the number of clusters beyond 4 does not significantly reduce the inertia.

Therefore, k=4 provides a good balance between model complexity and cluster separation.

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(df_market_data_transformed)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
k_4 = model.predict(df_market_data_transformed)

# Print the resulting array of cluster values.
print(k_4)
```

```python
# Create a copy of the scaled DataFrame
df_market_data_predictions = df_market_data_transformed.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_market_data_predictions['coin_segments'] = k_4

# Display the copy of the scaled DataFrame
df_market_data_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
df_market_data_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="coin_segments",
    hover_cols = "coin_id"
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
market_data_pca = pca.fit_transform(df_market_data_transformed)

# View the scaled PCA data
market_data_pca
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the first three principal components is approximately 89.5%.
This means that these three components together retain about 89.5% of the original data's variability, indicating that the dimensionality reduction preserved most of the important information.

```python
# Create a new DataFrame with the PCA data.
df_market_data_pca = pd.DataFrame(market_data_pca, columns=["PCA1", "PCA2", "PCA3"])

# Copy the crypto names from the original scaled DataFrame
df_market_data_pca["coin_id"] = df_market_data_transformed.index

# Set the coin_id column as index
df_market_data_pca = df_market_data_pca.set_index("coin_id")

# Display the scaled PCA DataFrame
df_market_data_pca.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_pca)
    inertia_pca.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia_pca}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_curve_pca = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
elbow_curve_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 
Based on the elbow method applied to the PCA-transformed data, the optimal number of clusters is k = 4. The inertia curve shows a clear bend at k=4, indicating that adding more clusters beyond this point yields diminishing returns in variance reduction.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** 
No, it does not differ. The best k value found using the original data was also 4. This consistency suggests that the dimensionality reduction preserved the underlying structure of the data and did not distort the cluster separability.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
k_4_pca = model.predict(df_market_data_pca)

# Print the resulting array of cluster values.
print(k_4_pca)
```

```python
# Create a copy of the scaled PCA DataFrame
df_market_data_pca_predictions = df_market_data_pca.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_market_data_pca_predictions["coin_segments"] = k_4_pca

# Display the copy of the scaled PCA DataFrame
df_market_data_pca_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
df_market_data_pca_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="coin_segments",
    hover_cols = "coin_id"
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
k_values = list(range(1, len(inertia) + 1))

df_elbow = pd.DataFrame({
    'k': k_values,
    'Original Data': inertia,
    'PCA Data': inertia_pca
})

df_elbow.hvplot.line(
    x='k',
    y=['Original Data', 'PCA Data'],
    title='Elbow Curve Comparison: Original vs PCA Data',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    line_width=2,
    legend='top_right'
)
```

```python
# Composite plot to contrast the clusters

# Predict cluster labels from KMeans models already fitted
# k_4 and k_4_pca are the predicted labels from KMeans on original and PCA data respectively
original_labels = k_4
pca_labels = k_4_pca

# Creating a DataFrame to hold PCA components and cluster labels from both models
# PCA1 and PCA2 are the first two principal components (used for 2D visualization)
df_clusters = pd.DataFrame({
    'PCA1': df_market_data_pca.iloc[:, 0],     # First principal component
    'PCA2': df_market_data_pca.iloc[:, 1],     # Second principal component
    'Original Cluster': original_labels,       # Cluster labels from original feature space
    'PCA Cluster': pca_labels                  # Cluster labels from PCA-transformed data
})

# Plot clusters based on original data (labels), visualized on PCA axes
original_cluster_plot = df_clusters.hvplot.scatter(
    x='PCA1',
    y='PCA2',
    c='Original Cluster',
    cmap='Category10',
    title='Clusters Based on Original Data',
    legend=False,
    size=70
)

# Plot clusters based on PCA-transformed data (labels), on same axes
pca_cluster_plot = df_clusters.hvplot.scatter(
    x='PCA1',
    y='PCA2',
    c='PCA Cluster',
    cmap='Category10',
    title='Clusters Based on PCA Data',
    legend=False,
    size=70
)

# Display both plots side by side for visual comparison
original_cluster_plot + pca_cluster_plot
```

#### Answer the following question: 

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:**  After visually analyzing the clustering results, using fewer features through PCA had little to no negative impact on the quality of the clustering. The clusters obtained from the PCA-transformed data were identical to those from the original dataset, suggesting that the core structure of the data was preserved. This indicates that dimensionality reduction using PCA can simplify the dataset, reduce computational complexity, and improve interpretability without sacrificing clustering performance, especially when a high percentage of variance (≈89.5%) is retained.
